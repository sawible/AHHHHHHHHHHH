```python
# importing packages
import astropy.constants as ac
import astropy.units as u
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Extract the first column and remove the first one entry.
Education_levels = np.array([
"Less than high school completion",
"High school completion",
"Some college, no degree",
"Associate’s degree",
"Bachelor’s degree",
"Master’s or higher degree"
])
```

```python
Mincome_arr = np.array([36700.,46400.,50200.,54200.,75100.,89200.,])
Fincome_arr = np.array([30900.,36700.,39100.,43100.,60800.,75700.,])
Education_arr = np.arange(len(Mincome_arr))
plt.plot(Education_arr,Mincome_arr,"o-")
plt.plot(Education_arr,Fincome_arr,"o-")
plt.xlabel("Education level")
plt.ylabel("Income")
plt.xticks(Education_arr,Education_levels,rotation=45,ha="right")
plt.show()
```

```python
# Use polyfit to fit the data with a linear relathioship, i.e., a first-order polynomial
# More information about polyfit can be found here: https://numpy.org/doc/stable/reference/generated/numpy.polyfit.html
zm_linear = np.polyfit(Education_arr,Mincome_arr,1)
# "income = z_linear[0] * level + z_linear[1]", where z_linear[0] represents the slope and z_linear[1] represents the y-intercept
print("Slope = ", zm_linear[0], " in dollar per level")
print("Y-intercept = ", zm_linear[1], " in dollar at level 0, i.e., Less than high school completion")
```

```python
# print out the slope equation for female data
zf_linear = np.polyfit(Education_arr,Fincome_arr,1)
print("Slope = ", zf_linear[0], " in dollar per level")
print("Y-intercept = ", zf_linear[1], " in dollar at level 0, i.e., Less than high school completion")
```

```python
zm_quadratic = np.polyfit(Education_arr,Mincome_arr,2)
print(zm_quadratic)

zf_quadratic = np.polyfit(Education_arr,Fincome_arr,2)
```

```python
mO = Mincome_arr
mO_err = np.array([3110, 1070, 2000, 4130, 4110, 8910])
mC_linear = zm_linear[0] * Education_arr + zm_linear[1]
mC_quadratic = zm_quadratic[0] * Education_arr**2 + zm_quadratic[1] * Education_arr + zm_quadratic[2]

fO = Fincome_arr
fO_err = np.array([3110, 960, 3070, 2580, 860, 4940])
fC_linear = zf_linear[0] * Education_arr + zf_linear[1]
fC_quadratic = zf_quadratic[0] * Education_arr**2 + zf_quadratic[1] * Education_arr + zf_quadratic[2]
```

```python
plt.errorbar(Education_arr, Mincome_arr, yerr=mO_err, marker="o", linestyle = "None", color = "navy", label = "Male Data")
plt.plot(Education_arr,mC_linear, color = "aqua", label = "Male Linear fit")
plt.plot(Education_arr,mC_quadratic, color = "blue", label = "Male Quadratic fit")

plt.errorbar(Education_arr, Fincome_arr, yerr=fO_err, marker="o", linestyle = "None", color = "purple", label = "Female Data")
plt.plot(Education_arr,fC_linear, color = "pink", label = "Female Linear fit")
plt.plot(Education_arr,fC_quadratic, color = "red", label = "Female Quadratic fit")
# i am sorry for reinforcing gender roles by making male blue and female pink
plt.legend()

plt.xticks(Education_arr,Education_levels,rotation=45,ha="right")
# please accept this graph as a token of my apology
plt.xlabel("Education Level")
plt.ylabel("Income [$]")
plt.show()
```
