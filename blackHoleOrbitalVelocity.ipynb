```python
import astropy.constants as ac
import astropy.units as u
import numpy as np
import matplotlib.pyplot as plt
```

How to define a function?

```python
def calcOrbitalVelocity(M, R):
  """
  This is a function that calculates the orbital velocity
  Input:
  M, the mass of the central object
  R, the orbtial radius
  Output:
  orbitalVelocity, the orbital velocity
  """
  orbitalVelocity = np.sqrt(ac.G * M / R)
  return orbitalVelocity
```

```python
# define input M and R
M = 1 * u.earthMass
R = 1 * u.earthRad

# calculate orbital velocity on Earth satellite
# using the predefined function
v_earthSatellite = calcOrbitalVelocity(M, R)
print(v_earthSatellite.to(u.km/u.s))
```

An Array: Group of numbers. Hundreds of cats, millions of dogs

```python
# orbital radius of the solar sustem
# mercury venus earth mars jupiter saturn uranus neptune
R_planets = np.array([0.4, 0.7, 1.0, 1.5, 5.2, 9.6, 19.2, 30.1]) * u.au
```

```python
# calculate orbital velocities for eight planets
M_sun = 1 * u.solMass
v_planets = calcOrbitalVelocity(M_sun, R_planets)
print(v_planets.to(u.km/u.s))
```

```python
plt.plot(R_planets.to(u.au), v_planets.to(u.km/u.s))
plt.xlabel("Orbital  (AU)".format(R_planets.unit))
plt.ylabel("Orbital Velocity (km/s)".format(v_planets.unit))
plt.show()
```

```python
M_hole = 4300000 * u.solMass # define the mass of the black hole as 4.3 million solar masses
print(M_hole) # print the mass of the black hole
print("")
R_parsec = (np.arange(1,31) * u.kpc) # define the array of radii in units of [parsecs]
print(R_parsec.to(u.km)) # checking to make sure
print("")
V_parsec = calcOrbitalVelocity(M_hole.to(u.solMass), R_parsec.to(u.km)) # calculate the orbital velocities for each radius
print(V_parsec.to(u.kpc/u.s)) # print the orbital velocities in units of [kiloparsecs/s]
```

```python
plt.plot(R_parsec.to(u.kpc), V_parsec.to(u.kpc/u.s))
plt.xlabel("Orbital Radius (KPC)".format(R_parsec.unit))
plt.ylabel("Orbital Velocity (KPC/s)".format(V_parsec.unit))
plt.show()
```
